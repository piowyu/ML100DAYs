```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.0001)
y_cos = np.cos(x)

plt.plot(x, y_cos)

plt.savefig("filename.png",dpi=300,format="png")

#畫出完整的 cos 圖形
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

T = np.arctan2(Y ,X )
plt.scatter(X,Y, c=T , alpha = .5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
